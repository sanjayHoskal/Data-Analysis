### 5. Convert datatype of column published_at from object to pandas datetime.

```python
df['published_at'] = pd.to_datetime(df['published_at'])        #using pandas to_datetime.
df['published_at']
```

### 6. Create a new column as 'published_month' using the column published_at (display the months only)

```python
def extract_month(date_str):
    return pd.to_datetime(date_str).strftime('%m')                      #using strftime inbuilt function along with %m string.

df['published_month'] = df['published_at'].apply(extract_month)
```

```python
df
```

###  7. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on.....

```python
def month(date_str):
    return pd.to_datetime(date_str).strftime('%b')          # using %b string which converts int(month) to its full form values.

df['published_month'] = df['published_at'].apply(month)

df
```

### 8. Find the number of videos published each month and arrange the months in a decreasing order based on the video count.

```python
video_counts = video_counts.sort_values(ascending=False)         #using ascending=false
video_counts
```

### 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.

```python
# Plot a scatter plot of views vs dislikes for each video
plt.scatter(df['view_count'], df['dislikes'], s=100)         
plt.xlabel('Views')
plt.ylabel('Dislikes')
plt.title('Views vs Dislikes')                    
plt.show()
```

```python
# some of the videos with minimum view count have high number of dislikes.
# Most viewed videos are comparable less dislikes.
# Most disliked videos are lies in minimum & medium viewed range.
```

```python
# Create a new column called views_vs_dislikes that is equal to dislikes divided by views
df['histogram'] = df['dislikes'] / df['view_count']

# Plot a histogram of the views_vs_dislikes column
plt.hist(df['histogram'])
plt.xlabel('Dislikes per view')
plt.ylabel('Number of views')
plt.show()
```
